```python
############################## velocity
cosine_sim_list = {}
pearson_list = {}
spearman_list = {}

for simuID in simuID_list:
    cosine_sim_list_tmp = {}
    pearson_list_tmp = {}
    spearman_list_tmp = {}
    
    #adata = scv.read("/lustre/home/zhangxiaochang/project/liyr/data/dyngen/result_0921/anndata/"+simuID+"/anndata.h5ad")
    
    
    for i in range(len(files)):
    #for i in [4]:
        print("\n")
        print(simuID +"_"+ files[i])
        print("\n")
        file2 = "/lustre/home/zhangxiaochang/project/liyr/data/dyngen/result_0921/velocity/"+simuID+"/"+files[i]+"/"+"anndata.h5ad"
        if os.path.exists(file2):        
            adata = scv.read(file2)
            #scv.tl.velocity_graph(adata)
            #scv.tl.velocity_embedding(adata,basis="pca")
            velocity = pd.DataFrame(adata.layers["velocity"])
            ground_truth_velocity = pd.DataFrame(adata.layers["ground_truth_velocity"].todense())
            
            # cosine similarity
            cosine_sim_matrix = cosine_similarity(velocity, ground_truth_velocity)
            cosine_sim = np.diag(cosine_sim_matrix)
            cosine_sim_list_tmp[files[i]] = np.median(cosine_sim)
        
            # pearson correlation
            pearson_cor = pd.DataFrame(ground_truth_velocity).corrwith(velocity, axis=1,method='pearson')
            pearson_list_tmp[files[i]] = np.median(pearson_cor.values)
        
            # spearman correlation
            spearman_cor = pd.DataFrame(ground_truth_velocity).corrwith(velocity, axis=1,method='spearman')
            spearman_list_tmp[files[i]] = np.median(spearman_cor.values)
        else:
            cosine_sim_list_tmp[files[i]] = np.nan
            pearson_list_tmp[files[i]] = np.nan
            spearman_list_tmp[files[i]] = np.nan
    cosine_sim_list[simuID] = cosine_sim_list_tmp
    pearson_list[simuID] = pearson_list_tmp
    spearman_list[simuID] = spearman_list_tmp
```

```python
pd.DataFrame(cosine_sim_list).to_csv("eval/cosine_sim_velocity.csv")
pd.DataFrame(pearson_list).to_csv("eval/pearson_sim_velocity.csv")
pd.DataFrame(spearman_list).to_csv("eval/spearman_sim_velocity.csv")
```
