```python
import scanpy as sc
import scvelo as scv
import pandas as pd
import numpy as np
import os
from itertools import product
```

```python
file_path = '/home/liyr/HuBMAP/Data/h5ad/'
file_list = os.listdir(file_path)
out_path = '/home/liyr/HuBMAP/RNA_velocity_result/scvelo/'

cluster = 'CytoTRACE2_Potency'
n_jobs = 28
t = 0
```

```python
for file in file_list :
    adata = sc.read_h5ad(file_path+file)
    series = adata.obs[cluster].value_counts()
    count_positive = (series > 0).sum()
    if count_positive > 1:
        t = t+1
        print('run number {},file_name is {}'.format(t,file))
        try :
            # process
            scv.pp.filter_and_normalize(adata,n_top_genes=2000)
            scv.pp.moments(adata)
            sc.pp.neighbors(adata)
            sc.tl.umap(adata)

            # run scvelo stochastic
            scv.tl.velocity(adata, mode='stochastic',n_jobs=n_jobs)

            adata.write_h5ad(out_path + file)

        except Exception as e:
            print('find error {}'.format(e))
```
