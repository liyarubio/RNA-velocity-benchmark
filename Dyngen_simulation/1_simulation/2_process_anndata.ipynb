```python
import numpy as np 
import os 
import pandas as pd 
from sklearn.metrics.pairwise import cosine_similarity 
from scipy import stats

SEED = 2024 
np.random.seed(SEED)
import seaborn as sns 
import matplotlib.pyplot as plt
import scvelo as scv
import scanpy as sc


os.chdir("/lustre/home/zhangxiaochang/project/liyr/data/dyngen/result_0921/")


#files = ['1_celldancer_pca', '2_scvelo_stochastic', '3_scvelo_dynamical', '4_unitvelo', '7_DeepveloGB','8_veloVI',"9_latentvelo","10_Dynamo_stochastic","10_Dynamo_deterministic","12_veloyto"]
#method_names = ['cellDancer', 'scvelo_stochastic', 'scvelo_dynamical', 'UniTvelo', 'DeepVelo_GB',"veloVI","LatentVelo","Dynamo_stochastic","Dynamo_deterministic","Velocyto"]


simuID_list = ['bifurcating_converging_seed1', 'bifurcating_converging_seed2', 'bifurcating_converging_seed3' ,'bifurcating_cycle_seed1', 'bifurcating_cycle_seed2', 'bifurcating_cycle_seed3' ,'bifurcating_loop_seed1', 'bifurcating_loop_seed2', 'bifurcating_loop_seed3' ,'bifurcating_seed1', 'bifurcating_seed2' ,'bifurcating_seed3' ,'binary_tree_seed1' ,'binary_tree_seed2', 'binary_tree_seed3', 'branching_seed1', 'branching_seed2', 'branching_seed3', 'consecutive_bifurcating_seed1', 'consecutive_bifurcating_seed2' ,'consecutive_bifurcating_seed3', 'converging_seed1', 'converging_seed2','converging_seed3', 'cycle_seed1', 'cycle_seed2', 'cycle_seed3' ,'cycle_simple_seed1' ,'cycle_simple_seed2', 'cycle_simple_seed3', 'disconnected_seed1' ,'disconnected_seed2', 'disconnected_seed3' ,'linear_seed1' ,'linear_seed2', 'linear_seed3', 'linear_simple_seed1' ,'linear_simple_seed2', 'linear_simple_seed3', 'trifurcating_seed1' ,'trifurcating_seed2' ,'trifurcating_seed3' ]
```

```python
for simuID in simuID_list:
    if not os.path.exists("anndata/"+simuID):
        os.makedirs("anndata/"+simuID)
    src_path = "/lustre/home/zhangxiaochang/project/liyr/data/dyngen/1_simulation_datasets/temporary_files/usecase_rna_velocity/datasets/"+simuID+"/anndata.h5ad"
    dst_path = "anndata/"+simuID
    #os.system(f"mkdir {dst_path}")
    adata = scv.read(src_path)
    # np.save("anndata/"+simuID+"/ground_truth_velocity.npy",adata.layers["ground_truth_velocity"])
    # print(adata)
    scv.pp.filter_and_normalize(adata, min_shared_counts=0, n_top_genes=2000)
    scv.pp.moments(adata, n_pcs=30, n_neighbors=30)
    if adata.obsm["X_pca"].max() > 10000:
        adata.obsm["X_pca"] = adata.obsm["X_pca"] / 1000
    scv.tl.umap(adata)
    if adata.obsm["X_umap"].max() > 10000:
        adata.obsm["X_umap"] = adata.obsm["X_umap"] / 1000
    sc.tl.leiden(adata,resolution=0.1)
    
    scv.tl.velocity(adata, mode='stochastic')
    adata.layers["velocity"] = adata.layers["ground_truth_velocity"].todense()
    scv.tl.velocity_graph(adata)
    scv.tl.velocity_embedding(adata,basis = "pca")
    scv.tl.velocity_embedding(adata, basis = "umap")
    adata.obsm["ground_truth_velocity_pca"] = adata.obsm["velocity_pca"]
    adata.obsm["ground_truth_velocity_umap"] = adata.obsm["velocity_umap"]
    adata.layers["velocity"] = adata.layers["ground_truth_velocity"]
    adata.write_h5ad(dst_path+"/anndata.h5ad")
```
