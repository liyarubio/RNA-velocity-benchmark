# young latentvelo result

```python
young_latent = sc.read_h5ad('/media/liyaru/LYR/Diff_change/11_redeem/adata/latentvelo.h5ad')
young_latent
```

```python
young_latent.obs_names
young2_HSPC.obs_names = [item.replace('_1', '') for item in young2_HSPC.obs_names]
```

# use intersect gene

```python
var_names = aged_latent.var_names.intersection(young_latent.var_names)
var_names
```

```python
young_latent2 = young_latent[:,var_names]
scv.tl.velocity_graph(young_latent2,vkey='velo',basis='umap',n_jobs=10)
scv.pl.velocity_embedding_stream(young_latent2,
                                 basis= 'umap',color="CellType",
                                 recompute=True,
                                 vkey="velo",
                                 legend_loc = 'right margin',
                                 groups = ['HSC','MPP','CMP','GMP','MEP','MKP','EryP',
                                           'MDP','Mono',
                                           'LMPP','CLP','ProB',
                                           'B','CD4','Plasma','cDC','pDC'],
                                 palette = celltype_palette,
                                 linewidth = 2,
                                 save="Young_overlap_gene.svg" 
                                 )
```
